```python
# An utility function to create dummy variable
def create_dummies( df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname)
    col_dummies.drop(col_dummies.columns[0], axis=1, inplace=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return df
```

```python
import statsmodels.formula.api as sm
from sklearn import metrics
```

```python
somersd_df=pd.DataFrame()
for num in df_new.columns.difference(['Churn']):
    logreg=sm.logit(str('Churn~')+str(num),data=df_new)
    result=logreg.fit()
    summ=result.summary()
    y_score=pd.DataFrame(result.predict())
    y_score.columns=['Score']
    somers_d=2*metrics.roc_auc_score(df_new['Churn'],y_score)-1
    temp=pd.DataFrame([num,somers_d]).T
    temp.columns=['Variable name','SomersD']
    somersd_df=pd.concat([somersd_df,temp],axis=0)
    
```

```python
names=['Logistic Regression','Decision Tree','RandomForest','AdaBoost',
       'KNN','SVC','Naive Bays']
classifiers=[LogisticRegression(),DecisionTreeClassifier(max_depth=9,max_features=9,min_samples_leaf=1,min_samples_split=4),
           RandomForestClassifier(max_depth=9,n_estimators=400),AdaBoostClassifier()
            ,KNeighborsClassifier(n_neighbors=3),SVC(),GaussianNB()]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score,average_precision_score,accuracy_score,confusion_matrix,f1_score
```

```python
for i,j in zip(names,classifiers):
    j.fit(train_X,train_y)
    
    
    score = j.score(test_X,test_y)
    y_pred =j.predict(test_X)
    cv_acc=cross_val_score(estimator=j,X=train_X,y=train_y,cv=5)
    cm=confusion_matrix(test_y,y_pred)
    cv_acc=cv_acc.mean()
    acc=accuracy_score(test_y,y_pred)
    f1=f1_score(test_y,y_pred)
    auc = roc_auc_score(test_y,y_pred)
    pre_sco=average_precision_score(test_y,y_pred)   

    
    
    print("+"*100)
    print("\n Classifier is \n" + str(i))
    print("\n Score \n" + str(score*100) + "%")
    print('\n Confusion Matrix \n',cm)
    print('\n Cross Validation Score \n',cv_acc)
    print('\n Testing Accuracy \n',acc)
    print('\n F1 Score \n',f1)
    print('\n AUC \n',auc)
    print('\n Precision Score \n',pre_sco)
    print("\n",j.get_params,"\n")
```
